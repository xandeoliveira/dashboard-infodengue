# Obtendo dados do Infodengue

Para buscar dados na API do Infodengue deve-se usar a seguinte url: `https://info.dengue.mat.br/api/alertcity?params` sendo que `params` é composto pelas variáveis: geocode, `disease`, `format`, `ew_start`, `ew_end`, `ey_start`, `ey_end`.

Mais detalhes no [tutorial oficial](https://info.dengue.mat.br/services/tutorial/Python) ou na [página oficial](https://info.dengue.mat.br/).

```python
import pandas as pd
```

```python
def url_factory(geocode, ew_start, ew_end, ey_start, ey_end):
    params = (
        "https://info.dengue.mat.br/api/alertcity?",
        "&disease=", "dengue",
        "&geocode=", geocode,
        "&disease=", "dengue",
        "&format=", "csv",
        "&ew_start=", ew_start,
        "&ew_end=", ew_end,
        "&ey_start=" , ey_start,
        "&ey_end=", ey_end    
    )
    
    return "".join(map(str, params))
```

```python
url_acarape_2024 = url_factory(2300150, 1, 54, 2024, 2024)
```

```python
df = pd.read_csv(url_acarape_2024, index_col='SE')
df.head(10)
```
